## 3. Clean data
Check for missing values and count total missing values from each column in descending order

```python
df.isnull().sum().sort_values(ascending=False)
```

## Comment
* license → 102,597 nulls out of 102,599. Basically empty => redundant
* house_rules → half missing => not really needed 
* country / country code will be the same since it's all NYC => not needed column
* 102599 entries and 26 columns
* price and service fees should be numbers instead of strings
* Because some rows have number of reviews equals 0 => its last review should also be NaN and the reviews per month should also show as 0 instead of NaN

```python
#1. Drop redundant columns
df.drop(columns=['country','country code', 'house_rules', 'license', 'host name'], inplace = True)
```

```python
#2. Drop rows with fewer than 50 missing values 
df.dropna(subset = ['lat', 'long', 'neighbourhood', 'neighbourhood group', 'cancellation_policy', 'instant_bookable', 'number of reviews', 'Construction year', 'price'], inplace = True)
```

```python
#3. Change data type of service fee and price
df['service fee']=df['service fee'].replace('[\$,]', "", regex=True).astype(float)
df['price']=df['price'].replace(r'[\$,]', "", regex=True).astype(float)
```

```python
#4. Change the reviews per month to 0 for listings with no reviews
df['reviews per month'] = df['reviews per month'].fillna(0)
```

```python
df.isnull().sum().sort_values(ascending=False)
```
